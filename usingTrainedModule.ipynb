```python
import cv2
import tensorflow as tf

CATEGORIES = ["Dog", "Cat"]


def prepare(filepath):
    IMG_SIZE = 150  # 50 in txt-based
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)


model = tf.keras.models.load_model("64x3-CNN.model")
DATADIR ="D:/Image Clustering/Image Cluster bw img/test1/17.jpg"

prediction = model.predict([prepare(DATADIR )])
print(prediction)  # will be a list in a list.
print(CATEGORIES[int(prediction[0][0])])
```

```python
import matplotlib.pyplot as plt 

DATADIR ="D:/Image Clustering/Image Cluster bw img/test1/100.jpg"
pic=plt.imread(DATADIR)

prediction = model.predict([prepare(DATADIR )])
print(prediction)  # will be a list in a list.
print(CATEGORIES[int(prediction[0][0])])

fig = plt.figure()
ax = fig.subplots() 
ax. imshow(pic) 
```

```python
import os
import cv2
import shutil
DATADIR = r"D:\Image Clustering\Image Cluster bw img\test1/"
for files in os.listdir (DATADIR):
    path = os.path.join(DATADIR,files)
    inputImage=cv2.imread(path)
    prediction = model.predict([prepare(path)])
    if (int(prediction[0][0])==1):
        cat=r"D:\Image Clustering\Image Cluster bw img\pridicted_cat/"
        outputpath=os.path.join(cat,files)
        cv2.imwrite(outputpath,inputImage)
    elif(int(prediction[0][0])==0):
        dog=r"D:\Image Clustering\Image Cluster bw img\pridicted_dog/"
        outputpath=os.path.join(dog,files)
        cv2.imwrite(outputpath,inputImage)
```
